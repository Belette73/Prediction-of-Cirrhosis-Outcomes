```python
df = pd.read_csv("train.csv",index_col=0)
df.Age = np.round((df.Age)/365,1)
```

```python
for i in ['Bilirubin','Cholesterol','Albumin', 'Copper','Alk_Phos',
       'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin']:
    if df[i].dtypes != "object":
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df["extreme_" + i] = df[i].apply(lambda x: 1 if (x < lower_bound or x > upper_bound) else 0)
```

```python
liste_name_colonne = []
liste_proportion_extreme = []
for i in df.columns:
    if "extreme_" in i:
        liste_name_colonne.append(i)
        proportion = df[i].sum()/len(df)
        liste_proportion_extreme.append(proportion)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Taille du graphique
plt.figure(figsize=(10, len(liste_name_colonne) * 0.5))

# Création du bar plot
ax = sns.barplot(x=liste_proportion_extreme, y=liste_name_colonne)

# Ajout de titre et labels
ax.set_title('Proportion de Valeurs Extrêmes par Feature')
ax.set_xlabel('Proportion')
ax.set_ylabel('Features')

# Formatage des valeurs en pourcentage et ajout des annotations
for p in ax.patches:
    width = p.get_width()
    plt.text(p.get_width(), p.get_y() + p.get_height() / 2, '{:1.2f}%'.format(width * 100), 
             va='center')

# Afficher le graphique
plt.show()
```
